```python

# Create a kernel density plot for 'Age' for passengers who did not survive (Survived == 0)
sns.kdeplot(data=df[df['Survived'] == 0]['Age'], label='Not Survived', shade=True)

# Create a kernel density plot for 'Age' for passengers who survived (Survived == 1)
sns.kdeplot(data=df[df['Survived'] == 1]['Age'], label='Survived', shade=True)

# Set labels and titles
plt.xlabel('Age')
plt.title('Kernel Density Plot of Age by Survival')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
